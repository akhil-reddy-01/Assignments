```python
import nltk
import gensim
import torch
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
df.tweet[1]
```

```python
length_new = list()
tweets=[]
for line in range(len(df)):
    review=[]
    for word in df.tweet[line].lower().split():
        if word not in set(stopwords.words('english')):
            review.append(word)
    review = ' '.join(review)
    tweets.append(review)
    length_new.append(len(review))
sns.distplot(length_new, kde=False, color='red', bins=100, )
plt.title('length of text', fontsize=18)
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(tweets)
testdataset=traindataset[int(t):]
traindataset=traindataset[0:int(t)]
train=df[:int(t)]
```

```python
train=df.iloc[:int(t),0]
```

```python
test=df.iloc[int(t):,0].values
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=100,criterion='entropy')
randomclassifier.fit(traindataset,train)
```

```python
test
```

```python
## Predict for the Test Dataset
test_transform= []

predictions = randomclassifier.predict(testdataset)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report

matrix=confusion_matrix(test,predictions)
print(matrix)
print('--------------------------------------------------')
score=accuracy_score(test,predictions)
print(score)
print('--------------------------------------------------')

report=classification_report(test,predictions)
print(report)
```
